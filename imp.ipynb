```python
import numpy as np
import pandas as pd
```

```python
class Imputer():
    def __init__(self,mode="mean",ext=False):
        self.mode=mode
        self.ext=ext
        self.repl={}
        self.nullCols=[];
    def fit_transform(self,df):

        self.nullCols=[col for col in df.select_dtypes(exclude=['object','datetime']).columns if (df[col].isnull().any())]
        for i in self.nullCols:
            self.repl[i]= df[i].mean() if self.mode=='mean'  else df[i].median() if self.mode=='median' else df[i].mode()[0]
        return self.transform(df)

    def transform(self,df_in):
        df=df_in.copy()
        for i in self.nullCols:
            for j in range(0,len(df[i])):
                if pd.isna(df.loc[j,i]):
                    df.loc[j,i]=self.repl[i]
        return df
```

```python
imp=Imputer()
imp_median=Imputer(mode='median')
imp_mode=Imputer(mode='mode')
```

```python
data=pd.read_csv('data.csv')
```
