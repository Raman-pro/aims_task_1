```python
import pandas as pd
```

```python
class OneHotEncoder():
    def __init__(self):
        self.colsToBeOHC={}
    def fit_transform(self,df):
        for i in df.select_dtypes(include=['object']).columns:
            if i not in self.colsToBeOHC:
                self.colsToBeOHC[i]=[i+"_"+uni for uni in df[i].unique()]
                print(self.colsToBeOHC)
        return self.transform(df)
    

    def transform(self,df_in):
        df=df_in.copy()
        for i in df.select_dtypes(include=['object']).columns:
            for j in range(len(df[i])):
                val=i+"_"+df.loc[j,i]
                if val in self.colsToBeOHC[i]:
                    df.loc[j,val]=1
        df=df.fillna(0)
        return df
```
