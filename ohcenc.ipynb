```python
class OneHotEncoder():
    def __init__(self):
        self.colsToBeOHC={}
    def fit_transform(self,df):
        for i in df:
            if i not in self.colsToBeOHC:
                self.colsToBeOHC[i]=df[i].unique()
        return self.transform(df)
    def transform(self,df_in):
        df=df_in.copy()
        for i in df:
            for j in range(len(df[i])):
                val=df.loc[j,i]
                if val in self.colsToBeOHC[i]:
                    df.loc[j,val]=1
        df=df.fillna(0)
        return df
```
