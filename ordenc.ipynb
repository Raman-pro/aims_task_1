```python
import pandas as pd
```

```python
class OrdinalEncoder():
    def __init__(self):
        self.colsEncoding=pd.DataFrame({})
    def fit_transform(self,df):
        for i in df.select_dtypes(include=['object']).columns:
            n=0
            for j in df[i]:
                if j not in self.colsEncoding.index or pd.isna(self.colsEncoding.loc[j, i]):
                    self.colsEncoding.loc[j,i]=n
                    n+=1
        print(self.colsEncoding)
        return self.transform(df)
    
    def transform(self,df_in):
        df=df_in.copy()
        for i in df.select_dtypes(include=['object']).columns:
            for j in range(len(df[i])):
                df.loc[j,i]=self.colsEncoding.loc[df[i][j],i]
        return df
                
            
```
